# Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Loading Data

```python
df = pd.read_csv('CC GENERAL.csv')
df.head()
```

```python
df.info()
```

# Exploratory Data Analysis

```python
df.describe()
```

```python
df.drop(['CUST_ID'],axis=1, inplace=True)
```

```python
df.isnull().sum()
```

## Checking for Skewnes

```python
plt.figure(figsize=(20,35))
for i, col in enumerate(df.columns):
    if df[col].dtype !='object':
        ax = plt.subplot(9, 2, i+1)
        sns.kdeplot(df[col],ax=ax)
        plt.xlabel(col)
        
plt.show()
```

```python
for col in df.columns:
    print(col)
    print({df[col].skew()})
```

```python
i=1
plt.figure(figsize=(20,40))
for col in df.columns:
    plt.subplot(9,2,i)
    df[col]= np.sqrt(df[col])
    sns.distplot(df[col])
    i=i+1
    
plt.show()
```

```python
plt.figure(figsize=(12, 12))
sns.heatmap(df.corr(), annot=True, cmap='crest')
```

## Dimensionality Reduction

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df = ss.fit_transform(df)

pca = PCA()
pca.fit(df)
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum())
```

```python
pca = PCA(n_components=6)
X = pca.fit_transform(df)
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
plt.figure(figsize=(15,10))
distortions = []
sil_scores = []
for i in range(2, 30):
    kmeans = KMeans(n_clusters=i, n_init=10, init= 'k-means++', algorithm='full', max_iter=300)
    kmeans.fit(X)
    distortions.append(kmeans.inertia_)
    label = kmeans.labels_
    sil_scores.append(silhouette_score(X, label))
plt.plot(np.arange(2,30,1), distortions, alpha=0.5)
plt.plot(np.arange(2,30,1), distortions, 'o', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(np.arange(2,30,1),sil_scores)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, n_init=10, init= 'k-means++', algorithm='full', max_iter=300)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
pca = PCA(n_components=2)
X2 = pca.fit_transform(df)
```

```python
pca_df = pd.DataFrame(data=X2, columns=['pca1', 'pca2'])
pca_df['labels'] = labels
pca_df.head()
```

```python
plt.figure(figsize=(10,10))
ax = sns.scatterplot(x='pca1', y='pca2', hue='labels', data=pca_df, palette='pastel')
```

```python
data_reduced = np.dot(df, pca.components_.T)
```

```python
data_original = np.dot(data_reduced,pca.components_)
```

```python
data_original
```
